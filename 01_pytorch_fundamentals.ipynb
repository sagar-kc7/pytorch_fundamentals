# **PyTorch Fundammentals**

```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print(torch.__version__)
```

## Introduction to Tensor

### *Creating Tensors*

```python
vector = torch.tensor([7, 7])
vector
```

```python
matrix = torch.tensor([[1, 3],
                       [3, 4]])
matrix
```

```python
matrix.ndim
```

```python
#Tensor

Tensor = torch.tensor([[[1,2,3],
                        [2,3,4],
                        [3,4,5]]])
Tensor
```

```python
Tensor[0]
```

### *Random Tensor*

```python
random_tensor = torch.rand(3,4)
random_tensor
```

```python
# Tensors of all Zeros

zeros = torch.zeros(size=(3,4))
zeros
```

```python
# Tensors of all ones

ones = torch.ones(size=(3,4))
ones
```

### Creating a range of Tensors

```python
one_to_ten = torch.arange(1, 11)
```

```python
# Tensor like

ten_zeros = torch.zeros_like(input=one_to_ten)
ten_zeros
```

### Tensor Datatype

```python
float_32_tensor = torch.tensor([3,4],
                               dtype=None)
float_32_tensor
```

```python
float_32_tensor.dtype

#it's int lets change to float32
```

```python
int_to_float32 = float_32_tensor.type(torch.float32)
int_to_float32.dtype
```

### Aggregation

```python
x = torch.arange(0, 100, 10)
x
```

```python
torch.max(x)
```

```python
torch.min(x)
```

```python
torch.mean(x.type(torch.float32))
```

### Stacking, Squeeze and Unsqueeze

```python
x = torch.zeros(2,1,2,1,2)
x, x.shape
```

```python
y = torch.squeeze(x)
y, y.shape
```

```python
y_unsqueezed = torch.unsqueeze(input=y, dim=0)
y_unsqueezed, y_unsqueezed.shape
```

### Indexing

```python
x[0]
```

```python
x[:]
```

```python
x[:, :, 1]
```
